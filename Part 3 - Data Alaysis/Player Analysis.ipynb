```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import csv
import requests
from PIL import Image, ImageDraw, ImageFont
import ast
from io import BytesIO
```

```python
df=pd.read_csv("../Common/Final Cleaned/player_df_cleaned.csv")
df.head(2)
```

# Single Variable Data Representation

```python
f, axes = plt.subplots(1, 3, figsize=(12, 6))
sb.histplot(df["trophies"], bins = 300, ax=axes[0])
sb.histplot(df["threeCrownWins"], bins = 300, ax=axes[1])
sb.histplot(df["expLevel"], bins = 45, ax=axes[2])

plt.show()
```

```python
df["arena"] = df["arena"].apply(lambda x: eval(x))
sb.countplot(y=df["arena"].map(lambda x: x["name"]), data=df)# Show the plot
plt.show()
```

# Multi-Variable Data Representations

```python
f, axes = plt.subplots(1, figsize=(12, 6))

sb.scatterplot(x="wins", y="trophies", data=df, ax=axes)
axes.set_title("Scatter plot of wins and trophies")
```

```python
import csv
import os
import requests
import ast

# Define the path to your CSV file
csv_path = "card_info.csv"

# Define the folder path where you want to store the images
folder_path = "c"

# Create the folder if it doesn't exist
if not os.path.exists(folder_path):
    os.makedirs(folder_path)

# Open the CSV file
with open(csv_path, "r") as csv_file:
    csv_reader = csv.DictReader(csv_file)

    # Loop through each row in the CSV file
    for row in csv_reader:

        # Get the image URL from the "iconUrls" column
        cardURL = ast.literal_eval(row["iconUrls"])
        image_url = cardURL["medium"]

        # Get the image name from the "name" column
        image_name = row["name"] + ".png"

        # Define the full path where you want to store the image
        image_path = os.path.join(folder_path, image_name)

        # Download the image using the requests module
        response = requests.get(image_url)
        with open(image_path, "wb") as image_file:
            image_file.write(response.content)

        print(f"Downloaded {image_name}")
```
